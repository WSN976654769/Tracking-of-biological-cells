```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2

tf.get_logger().setLevel('ERROR')
tf.get_logger().warning('test')

imgset_no = 3

if (imgset_no == 1):
    imgset = './data/DIC-C2DH-HeLa'
    tf_model_path = "./model_1/"
    w, h =  256, 256
    
if (imgset_no == 3):
    imgset = './data/PhC-C2DL-PSC'
    tf_model_path = "./model_3/"
    w, h =  720, 576

# data
img_path = imgset + "/Sequence 2/t081.tif"
print(img_path)

img = cv2.imread(img_path, 0)
plt.imshow(img)
plt.show()

# prepare_and_predict
img = cv2.resize(img, (w,h))
print(img.shape)

X = np.reshape(img, (1, h, w, 1))
print(X.shape)
X = (X/255).astype(np.float32)

# model
loaded = tf.saved_model.load(tf_model_path)
npy_tensor = loaded.signatures['serving_default'](tf.convert_to_tensor(X))['output'].numpy()
npy = np.reshape(npy_tensor, (h, w))* 255 
npy = np.array(npy, dtype="u1")

# output image
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")

plt.subplot(1,2,2)
plt.imshow(npy, 'gray')
plt.show()
```

```python
from scipy import ndimage as ndi
from skimage.segmentation import watershed
from skimage.feature import peak_local_max
from sklearn.cluster import MeanShift
from skimage.morphology import h_maxima



if (imgset_no == 1):
    
    print("1")
    plt.figure(figsize=(14,7))
    
    print("GaussianBlur -- H-maxima -- Opening -- Find contours")
    img_blur = cv2.GaussianBlur(npy, (15,15),0)
    plt.subplot(1,3,1)
    plt.imshow(img_blur, cmap='gray')

    img_h = h_maxima(img_blur, 20, selem=None)
    kernel = np.ones((7,7),np.uint8)
    img_h_opening = cv2.dilate(img_h, kernel)
    plt.subplot(1,3,2)
    plt.imshow(img_h_opening, cmap='gray')
    
    img_show = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    image, contours, hierarchy = cv2.findContours(img_h_opening,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(img_show,(x,y),(x+w,y+h),(0,255,0),2)
    
    plt.subplot(1,3,3)
    plt.imshow(img_show, cmap='gray')
    plt.show()
    
    print(len(contours))

if (imgset_no == 3):
    
    print("3")
    plt.figure(figsize=(10,5))
    
    plt.subplot(1,2,1)
    plt.imshow(img, cmap="gray")

    kernel = np.ones((9,9),np.uint8)
    img_tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
    thresh = cv2.threshold(img_tophat, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

    plt.subplot(1,2,2)
    plt.imshow(thresh, cmap='gray')
    plt.show()
    
    plt.figure(figsize=(14,7))
    
    print("Threshold -- Opening -- Watershed")
    thresh = cv2.threshold(npy, 127, 255, cv2.THRESH_BINARY)[1]
    plt.subplot(1,3,1)
    plt.imshow(thresh, cmap='gray')
    
    kernel = np.ones((3,3),np.uint8)
    opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN, kernel, iterations = 2)
    plt.subplot(1,3,2)
    plt.imshow(opening, cmap='gray')

    distance = ndi.distance_transform_edt(opening)
    local_max = peak_local_max(distance, indices=False, min_distance=10, labels=opening)
    markers = ndi.label(local_max, structure=np.ones((3, 3)))[0]
    ws_labels = watershed(-distance, markers=markers, mask=opening)
    ret, ws_labels= cv2.threshold(ws_labels.astype(np.uint8), 5, 255, cv2.THRESH_BINARY)
    
    img_show = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    image, contours, hierarchy = cv2.findContours(ws_labels,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    
    plt.subplot(1,3,3)
    plt.imshow(img)
    plt.show()
    
    print(len(set(ws_labels.flatten()))-1)
      
```
